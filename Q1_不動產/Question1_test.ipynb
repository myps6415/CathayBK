```python
# 使用【pandas】套件
import pandas as pd

'''
讀取檔名【 a_lvr_land_a 】【 b_lvr_land_a 】 【 e_lvr_land_a 】【 f_lvr_land_a 】 【 h_lvr_land_a 】五份資料集
建立 dataframe 物件【df_a】【df_b】【df_e】【df_f】 【df_h】
'''

df_a = pd.read_csv('/Users/johntung/Downloads/cathay/a_lvr_land_a.csv')
df_b = pd.read_csv('/Users/johntung/Downloads/cathay/b_lvr_land_a.csv')
df_e = pd.read_csv('/Users/johntung/Downloads/cathay/e_lvr_land_a.csv')
df_f = pd.read_csv('/Users/johntung/Downloads/cathay/f_lvr_land_a.csv')
df_h = pd.read_csv('/Users/johntung/Downloads/cathay/h_lvr_land_a.csv')

df_a.head()
```

```python
# 操作 dataframe 物件，將五個物件合併成【df_all】
df_all = pd.concat([df_a, df_b, df_e, df_f, df_h])
print(len(df_all))
df_all.head()
```

```python
df_all = df_all.drop(0).reset_index(drop=True)
print(len(df_all))
df_all.head()
```

```python
'''
以下列條件從【df_all】篩選/計算出結果，並分別輸出【csv 檔案】 :
* filter_a.csv
  -【主要用途】為【住家用】
  -【建物型態】為【住宅大樓】
  -【總樓層數】需【大於等於十三層】
'''

# 因總樓層數為國字，需轉為阿拉伯數字去篩選大於等於十三層的物件
df_all['總樓層數'].unique()
```

```python
# drop nan
filter_a = df_all.dropna(subset=['總樓層數']).reset_index(drop=True)
print(len(filter_a))
filter_a.head()
```

```python
num_dict = {'一':1, '二':2, '三':3, '四':4, '五':5, '六':6, '七':7, '八':8, '九':9, '十':10}

filter_a['總樓層數'] = [i.replace('層', '') for i in filter_a['總樓層數']]
filter_a['總樓層數'].unique()
```

```python
# 將中文數字轉阿拉伯數字
floor_list = []

for chi_num in filter_a['總樓層數']:
    if '十' in chi_num:
        if len(chi_num) == 1:
            floor_list.append(num_dict[chi_num])
        elif '十' in chi_num[0]:
            num = 0
            for i in chi_num:
                num += num_dict[i]
            floor_list.append(num)
        else:
            if len(list(chi_num)) == 2:
                floor_list.append(num_dict[chi_num[0]] * num_dict[chi_num[1]])
            else:
                floor_list.append(num_dict[chi_num[0]] * num_dict[chi_num[1]] + num_dict[chi_num[2]])
    else:
        floor_list.append(num_dict[chi_num])
        
print(len(floor_list))
print(floor_list)
```

```python
filter_a['總樓層數_num'] = floor_list
print(len(filter_a))
filter_a.head()
```

```python
filter_a[['總樓層數', '總樓層數_num']]
```

```python
'''
以下列條件從【df_all】篩選/計算出結果，並分別輸出【csv 檔案】 :
* filter_a.csv
  -【主要用途】為【住家用】
  -【建物型態】為【住宅大樓】
  -【總樓層數】需【大於等於十三層】
'''
filter_a[(filter_a['主要用途'] == '住家用') & (filter_a['建物型態'].str.contains('住宅大樓')) & (filter_a['總樓層數_num'] >= 13)].to_csv('filter_a.csv', index=False)
```

```python
'''
* filter_b.csv
- 計算【總件數】
- 計算【總車位數】(透過交易筆棟數)
- 計算【平均總價元】
- 計算【平均車位總價元】
'''
df_all.columns
```

```python
df_all.info()
```

```python
df_all['總價元'] = df_all['總價元'].astype(int)
df_all['車位總價元'] = df_all['車位總價元'].astype(int)
```

```python
'''
* filter_b.csv
- 計算【總件數】
- 計算【總車位數】(透過交易筆棟數)
- 計算【平均總價元】
- 計算【平均車位總價元】
'''

df_all['Parking_space'] = [int(i.split('車位')[1]) for i in df_all['交易筆棟數']]

total_num = len(df_all)
pk_space = sum(df_all['Parking_space'])
average_yen = int(sum(df_all['總價元']) / total_num)
parking_space_yen = int(sum(df_all['車位總價元'])/ pk_space)
print(total_num, pk_space, average_yen, parking_space_yen)
```

```python
filter_b = pd.DataFrame({'總件數':[total_num], '總車位數':[pk_space], '平均總價元':[average_yen], '平均車位總價元':[parking_space_yen]})
filter_b
```

```python
filter_b.to_csv('filter_b.csv', index=False)
```
